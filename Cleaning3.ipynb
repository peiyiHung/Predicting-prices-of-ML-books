```python
from fastai.tabular.all import *
```

```python
books = pd.read_csv("bookprice_cleaned_v2.csv")
```

```python
terms = ["machine learning", "statistical learning", 
         "deep learning", "reinforcement learning",
         "machine-learning", "natural language processing",
         "semi-supervised learning", "algorithms"]


notML = []
for i, n in enumerate(books.Name):
    isML = False
    for t in terms:
        if t in n.lower():
            isML = True
    if not isML:
        notML.append(i)
        print(i, n)
```

```python
to_keep = [23, 26, 51,57, 29, 490, 910, 991, 1209, 1260]
```

```python
for i in notML:
    if i not in to_keep:
        books.drop(i, inplace=True)
```
