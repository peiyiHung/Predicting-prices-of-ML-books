## Number of Pages

```python
books["NumberOfPages"].describe()
```

```python
drop_idx = []
for i in books[books["NumberOfPages"]<50].index:
    if "machine learning" not in books.loc[i, "Name"].lower():
        drop_idx.append(i)
```

```python
books.drop(drop_idx, inplace=True)
```

```python
books["NumberOfPages"].describe()
```

```python
books[books["NumberOfPages"]<10]
```

## Date

```python
books["Publication date"] = books["Publication date"].astype("datetime64")
```

```python
books["Year"] = books["Publication date"].apply(lambda x:x.year)
```

```python
books["Year"].value_counts()
```

```python
books = books[books["Year"] > 2004]
```

```python
books["Month"] = books["Publication date"].apply(lambda x:x.month)
```

```python
books["Month"].value_counts()
```
